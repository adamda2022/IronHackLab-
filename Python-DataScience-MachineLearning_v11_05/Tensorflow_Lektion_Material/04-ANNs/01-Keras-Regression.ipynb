#### Featurespalten
    
* id - Einzigartige ID der verkauften Wohnung
* date - Datum des Verkaufs
* price - Preis der verkauften Wohnung
* bedrooms - Anzahl der Schlafzimmer
* bathrooms - Anzahl von Badezimmer, wobei 0.5 ein Badezimmer mit Toilette, aber ohne Dusche bezeichnet
* sqft_living - Quadratfuss des inneren Wohnraums
* sqft_lot - Quadratfuss Grundstücksgröße
* floors - Anzahl Etagen
* waterfront - Bezeichnet ob die Wohnung einen Blick auf eine Wasserfläche hat oder nicht
* view - Ein Index von 0 bis 4 über die Qualität der Aussicht der Wohnung
* condition - Ein Index von 1 bis 5 über den Zustand der Wohnung
* grade - Ein Index von 1 bis 13: 1-3 bedeutet schlechte Konstruktion und Design, 7 durschnittliche Konstruktion und Design und 11-13 exzellente Konstruktion und Design
* sqft_above - Quadratfuss Wohnfläche überebenerdig 
* sqft_basement - Quadratfuss Wohnfläche unterebenerdig 
* yr_built - Baujahr des Hauses
* yr_renovated - Jahr der letzten Rennovierung
* zipcode - Postleitzahl des Hauses
* lat - Breitengrad
* long - Längengrad
* sqft_living15 - Quadratfuss der Wohnfläche der 15 nächsten Nachbarn
* sqft_lot15 - Quadratfuss der Grundstücksgröße der 15 nächsten Nachbarn

## Skalierung und Test-/Trainingaufteilung

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

### Skalierung

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Modell erzeugen

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

## Modell trainieren

```python
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

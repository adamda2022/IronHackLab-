```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
import numpy as np
```

```python
url = "https://www.billboard.com/charts/hot-100/"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
songname=soup.select("#title-of-a-story.a-no-trucate")
#songname
```

```python
name_of_song=[]
for i in songname:
    name_of_song.append(i.get_text(strip=True))
```

```python
artist_name= soup.select(".c-label.a-no-trucate")
#artist_name
```

soup.select("#title-of-a-story.a-no-trucate")[1].get_text(strip=True)

```python
name_of_artist=[]
for i in artist_name:
    name_of_artist.append(i.get_text(strip=True))
#name_of_artist
```

```python
songs = pd.DataFrame({"name_of_artist":name_of_artist,
                       "name_of_song":name_of_song,
                      })

songs['name_of_song']
```

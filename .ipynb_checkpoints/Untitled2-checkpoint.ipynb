```python
df_nummerical= df.select_dtypes(np.number)
df_nummerical
```

```python
def Remove_Outlier_Indices(df_nummerical):
    Q1 = df_nummerical.quantile(0.25)
    Q3 = df_nummerical.quantile(0.75)
    IQR = Q3 - Q1
    trueList = ~((df_nummerical < (Q1 - 1.5 * IQR)) |(df_nummerical > (Q3 + 1.5 * IQR)))
    return trueList

indices = Remove_Outlier_Indices(df_nummerical)
df_trim = df[indices.all(1)]
```
